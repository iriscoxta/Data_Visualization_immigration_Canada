```python
import pandas as pd
```

```python
df = pd.read_csv("/content/imigrantes_canada.csv")
```

```python
df.set_index("País", inplace=True)
```

```python
# criando variavel para guardar intervalo de anos
anos = list(map(str, range(1980,2014)))
```

```python
anos
```

```python
# pegando somente a parte do Brasil
brasil = df.loc["Brasil", anos]
```

```python
brasil
```

```python
brasil_dict = {
    'ano' : brasil.index.tolist(),
    'imigrantes' : brasil.values.tolist()
}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
# no eixo X ficam os dados independentes e no Y os dados dependentes, por exemplo, com os passar dos anos (independente - x) o numero de imigrantes (dependente - y) vai mudando
plt.plot(dados_brasil["ano"], dados_brasil["imigrantes"])
plt.title("Imigrantes do Brasil para o Canadá")
plt.xlabel("Ano")
plt.ylabel("Número de imigrantes")
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2010"])

# plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show() # tudo que for escrito depois dessa linha não vai ser executado pois plt.show() impede
```

```python
# criando figura com 4 graficos
fig, axs = plt.subplots(2, 2, figsize=(15, 5))
# ajusta altura (h) e largura (w) dos graficos todos simultaneos
fig.subplots_adjust(hspace=0.5, wspace=0.3)
# adicionando titulo geral
fig.suptitle("Imigração dos quatros países da América do Sul para o Canadá 1980 - 2013 \n", fontsize=14)

#[linha, coluna]
axs[0,0].plot(df.loc["Brasil", anos])
axs[0,0].set_title("Brasil")

axs[0,1].plot(df.loc["Colômbia", anos])
axs[0,1].set_title("Colômbia")

axs[1,0].plot(df.loc["Argentina", anos])
axs[1,0].set_title("Argentina")

axs[1,1].plot(df.loc["Peru", anos])
axs[1,1].set_title("Peru")

#criando um loop para percorrer todos os subplots nessa matriz
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel("Ano", fontsize=10)
  ax.set_ylabel("Número de imigrantes", fontsize=10)
  ax.xaxis.set_tick_params(labelsize=12)
  ax.yaxis.set_tick_params(labelsize=12)

# ajustando valores do eixo y
ymin = 0
ymax = 7000

# .ravel é um método que consegue transformar um array bidimensional em um array unidimensional
for ax in axs.ravel():
  # limites do eixo y
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
# ordendando o dataframe dcom os maiores valores ate os menores
america_sul_sorted = america_sul.sort_values("Total", ascending=True)
```

```python
cores = ["royalblue", "orange", "forestgreen", "orchid",
         "purple", "brown", "slateblue", "gray", "olive",
         "navy", "teal", "tomato"]



fig, ax = plt.subplots(figsize=(12,5))
# barh = barras na horizontal
ax.barh(america_sul_sorted.index, america_sul_sorted["Total"], color=cores)
ax.set_title("Imigração da América do Sul para o Canadá \n 1980 - 2013", fontsize=18, loc="left")
ax.set_xlabel("Número de imigrantes", fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
# criando um gráfico para destacar o Brasil
cores = []
for pais in america_sul_sorted.index:
  if pais == "Brasil":
    cores.append("green")
  else:
      cores.append("silver")

fig, ax = plt.subplots(figsize=(12,5))
# barh = barras na horizontal
ax.barh(america_sul_sorted.index, america_sul_sorted["Total"], color=cores)
ax.set_title("Imigração da América do Sul: Brasil foi 4º país com mais imigrantes \n para o Canadá de 1980 a 2013", fontsize=18, loc="left")
ax.set_xlabel("Número de imigrantes", fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)


#adicionando anotações nas barras
for i, v in enumerate(america_sul_sorted["Total"]):
  # ax.text => add texto / v + 20 => indica posição de onde o texto vai ser add (v = valor das barras) / i = posição do texto /
  # str(v) => converte o tamanho de cada barra em uma string (texto) / ha = alinhamento horizontal / va = define alinhamento vertical
  ax.text(v + 20, i, str(v), color="black", fontsize=10, ha="left", va="center")

# remove caixa que envolve grafico
ax.set_frame_on(False)

# remove eixo y
ax.get_xaxis().set_visible(False)

# remove grid
ax.grid(False)

# remove marcações de ticks / axis = define quais eixos ,  which = quais ticks queremos retirar , lenght = comprimento dos marcadores
ax.tick_params(axis="both", which="both", length=0)

# salvando
fig.savefig("imigracao_america_sul.png", transparent=False, dpi=300, bbox_inches="tight")

plt.show()
```

```python
# criando uma função para gerar o gráfico, essa função vai pedir o nome da paleta de cores

def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y=top_10.index, x="Total", orient="h", palette=palette)

  ax.set_title("Países com maior imigração para o Canadá\n 1980 a 2013", fontsize=16, loc="left")
  ax.set_xlabel("Número de imigrantes", fontsize=12)
  ax.set_ylabel("")

  plt.show()
```

# **Escolhendo paletas de cores**

1.    Considerar o tipo de dado que está sendo plotado: dados categóricos, dados sequenciais ou dados divergentes requerem diferentes tipos de paletas de cores, como discutimos no vídeo anterior.
2.   Escolha cores que sejam facilmente distinguíveis umas das outras, se você estiver usando várias séries de dados em um único gráfico.

3.   Evite cores muito brilhantes e saturadas, pois podem ser muito intensas e distrair a atenção do público.
4.    Use cores de maneira consistente em todos os seus gráficos para facilitar a comparação entre eles.

5.   Considere o contexto em que o gráfico será utilizado e se as cores escolhidas são apropriadas para esse contexto.








## **Paletas de cores sequenciais**

```python
gerar_grafico_paleta("Blues")
```

```python
# o 'r' é usado pra inverter a ordem da paleta
gerar_grafico_paleta("Blues_r")
```

## **Paletas de cores divergentes**

```python
gerar_grafico_paleta("rocket")
```

```python
# usar com dados que estão variando em torno de um valor central
gerar_grafico_paleta("coolwarm")
```

## **Paletas de cores categóricas**

```python
gerar_grafico_paleta("tab10")
```

```python
# criando animação
dados_brasil['ano'] = dados_brasil['ano'].astype(int)
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[dados_brasil['ano'].iloc[0]], y=[dados_brasil['imigrantes'].iloc[0]], mode='lines', name='Imigrantes', line=dict(width=4))
)

# Definindo as configurações de layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir as configurações de animação
frames = [go.Frame(data=[go.Scatter(x=dados_brasil['ano'].iloc[:i+1], y=dados_brasil['imigrantes'].iloc[:i+1])]) for i in range(len(dados_brasil))]
fig.frames = frames

# Mostrando a figura
fig.show()
```
